```python
# question 7

# import image module from PIL library to get acces to image funtionality in google colab notebooks

from PIL import Image
img = Image.open('/content/drive/MyDrive/photo_2024-08-08 23.39.48.jpeg')
display(img)

# this is the image of the flowchart for the sieve of eratosthenes algorithm


# import math library for the ceil() function
import math

# factorial function using while loop
def factorial(x):
  ans=1;
  while(x):
    ans*=x
    x-=1
  return ans

# prime array initialised with 101 ones ie upto 100 index
# will finally contain 1 at prime indices and 0 at non prime indices
prime = [1] * 101


# this program is using sieve of eratosthenes concept
# in which to check whether a number is prime or not , we check its divisibility by all known primes less than sqaure root of the number
# if it is divisible by some prime, we change the value in the prime list at that index from 1 to 0, indicating that the number is not prime

# intial condition to start iteration
prime[1]=0
prime[0]=0

# implementation of above mentioned sieve of eratosthenes concept
for i in range(4,101):
  for j in range(2,math.ceil(math.sqrt(i))+1):
    if(prime[j]==1 and i%j==0):
      prime[i]=0
      break

# list of primes
print(f"\n\n\n\nList of primes:")
for i in range(101):
  if(prime[i]):
    print(i, end=' ')

# list of factorials of primes
factorial_list = []
print(f"\n\nList of factorials of primes:")
for i in range(101):
  if(prime[i]):
    factorial_list.append(factorial(i))
    print(factorial_list[-1],end=' ')
```

```python
# question 3 (continuing from question 1)

# NOTE: We drop the 'Year' and 'Month' columns as they
#       1. Create unnecessarily many NaN entries in the mode dataframe in other columns
#       2. Calculating any of statistical quantities for year and month is not sensible as they are the independent variables
#       3. We are rather more interested in the statistical quantities of the anomalies and uncertainities in different time series
# so everytime we first drop these columns using the drop() method

# get mean, median, mode of the dataframe using the corresponding methods like

# mean()
print(f"The mean of dataframe is (column/header-wise)\n")
print(df.drop(['Year','Month'], axis=1).mean())
print(f"\n\n\n")

# median()
print(f"The median of the dataframe is (column/header-wise)\n")
print(df.drop(['Year','Month'], axis=1).median())
print(f"\n\n\n")


# mode()
print(f"The mode of the dataframe is (column/header-wise)\n")
print(df.drop(['Year','Month'], axis=1).mode())
print(f"\n\n\n")


# we can get the standard deviation from getting the row indexed "std" in the returned dataframe
# of the describe() method by using the loc() method and further rename it using the rename() method
print(f"The standard deviation of the dataframe is (column/header-wise)\n")
print(df.drop(['Year','Month'], axis=1).describe().loc['std'].rename('standard_deviation'))
print(f"\n\n\n")


# as variance is the square of standard deviation, we define a square function
def square(x):
  return x*x

# we declare a new pandas series, "variance_series" by getting the "std" indexed row from the describe() method on the
# original dataframe, rename it using the rename() method, and square the entries using the apply() method with the square function
# passed as an argument
print(f"The variance of the dataframe is (column/header-wise)\n")
variance_series = df.drop(['Year','Month'], axis=1).describe().loc['std'].rename("variance").apply(square)
print(variance_series)
```
