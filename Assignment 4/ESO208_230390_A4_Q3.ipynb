```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
# question 3 subpart ii

# ignoring A as it is only a constant scaling factor

# the gaussian function
def f(x,a):
  return np.exp(-a*(x**2))

# the fourier transform of the gaussian found in part i in the handwritten solutions
def c(k,a):
  return np.exp(-(k**2)/(4*a))/(2*np.sqrt(np.pi * a))

# vectorized versions of the above functions to accept array inputs
v_f = np.vectorize(f)
v_c = np.vectorize(c)
```

```python
# declare the vector arguments for the plotting commands
n = 1000
x = np.linspace(-10,10,num = n,endpoint=True)
a1 = np.full(n,0.5)
a2 = np.full(n,1)
a3 = np.full(n,2)
a4 = np.full(n,10)
```

```python
# plotting commands
fig = plt.figure()
fig.set_figwidth(12)
fig.set_figheight(6)

plt.plot(x,v_f(x,a1))
plt.plot(x,v_f(x,a2))
plt.plot(x,v_f(x,a3))
plt.plot(x,v_f(x,a4))
plt.legend(['a = 0.5','a = 1','a = 2','a = 10'])
plt.title("Gaussian Plots for Different 'a'")
plt.ylabel("f(x)")
plt.xlabel("x")
plt.show()
```

It can be seen clearly from the above plot that, as the value of 'a' increases the width of the Gaussian plot decreases but the maximum value (height) remains unchanged.

```python
# declare the vector arguments for the plotting commands
n = 2000
x = np.linspace(-20,20,num = n,endpoint=True)
a1 = np.full(n,0.5)
a2 = np.full(n,1)
a3 = np.full(n,2)
a4 = np.full(n,10)
```

```python
# plotting commands
fig = plt.figure()
fig.set_figwidth(12)
fig.set_figheight(6)

plt.plot(x,v_c(x,a1))
plt.plot(x,v_c(x,a2))
plt.plot(x,v_c(x,a3))
plt.plot(x,v_c(x,a4))
plt.legend(['a = 0.5','a = 1','a = 2','a = 10'])
plt.title("Frequency Spectrum for Different 'a'")
plt.ylabel("c(k)")
plt.xlabel("k")
plt.show()
```

It can be seen clearly from the above plot that, as the value of 'a' increases the width of the spectrum also increases whereas the height of the spectrum decreases.
