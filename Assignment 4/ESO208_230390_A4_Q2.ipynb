```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
```

```python
# question 2 subpart ii

# according to the fourier series found in part i in the handwritten solutions

# the sawtooth function can be expressed as (-AL/π) * Σ(k=1 to ∞) (-1)^k / k * sin(2πkx/L)

# using A = 1 and L = 1 as given in the question

# define the function saw_tooth_gen
def saw_tooth_gen(x,n):

  # input:
  # n = number of terms from the fourier series to be included in the summation

  ans = 0.0

  for k in range(1,n+1):
    ans += pow(-1,k)* np.sin(2*np.pi*k*x) / k

  return -(ans/np.pi)

# vectorize the function to take vector inputs
v_stg = np.vectorize(saw_tooth_gen)
```

```python
# define arrays for plotting commands
n = 1000
x = np.linspace(-2,2,num = n,endpoint=True)
n1 = np.full(n,1)
n2 = np.full(n,5)
n3 = np.full(n,20)
n4 = np.full(n,100)
n5 = np.full(n,250)
```

```python
# number of terms = 1
plt.plot(x,v_stg(x,n1))
plt.plot(x,np.zeros(n))
plt.xlabel("x")
plt.ylabel('f(x)')
plt.title('Fourier Series Approimation of Saw-Tooth Function with 1 term')
plt.show()
```

```python
# number of terms = 5
plt.plot(x,v_stg(x,n2))
plt.plot(x,np.zeros(n))
plt.xlabel("x")
plt.ylabel('f(x)')
plt.title('Fourier Series Approimation of Saw-Tooth Function with 5 terms')
plt.show()
```

```python
# number of terms = 20
plt.plot(x,v_stg(x,n3))
plt.plot(x,np.zeros(n))
plt.xlabel("x")
plt.ylabel('f(x)')
plt.title('Fourier Series Approimation of Saw-Tooth Function with 20 terms')
plt.show()
```

```python
# number of terms = 100
plt.plot(x,v_stg(x,n4))
plt.plot(x,np.zeros(n))
plt.xlabel("x")
plt.ylabel('f(x)')
plt.title('Fourier Series Approimation of Saw-Tooth Function with 100 terms')
plt.show()
```

```python
# number of terms = 250
plt.plot(x,v_stg(x,n5))
plt.plot(x,np.zeros(n))
plt.xlabel("x")
plt.ylabel('f(x)')
plt.title('Fourier Series Approimation of Saw-Tooth Function with 250 terms')
plt.show()
```

As the number of terms used from the fourier series is increased, the representation becomes more faithful to the original function whose fourier series is being computed.
